# Question 1 - Find how many words and how many unique wwords in file 'austen-persuasion.txt'

```python
import nltk
from nltk.corpus import gutenberg #Importing the gutenberg 
gutenberg.fileids() 
string = len(gutenberg.words('austen-persuasion.txt')) 
print("Number of words in the file: ",string)
string1 = len(set(gutenberg.words('austen-persuasion.txt')))
print("Number of unique words in the file: ",string1)
```

# Calculate the frequency for each words

```python
import nltk
from nltk.corpus import gutenberg
#from collections import Counter
words=nltk.corpus.gutenberg.words('austen-persuasion.txt')
d=dict()
for c in words:
    if c not in d:
        d[c]=1
    else:
        d[c]=d[c]+1
print(d)
```

# show the top 10 most frequent words and their frequency

```python
popular_words = sorted(d.items(), key = lambda x : x[1] ,reverse = True) 
top_10 = popular_words[:10] #Top 10 popular words from popular_words list
print('The top 10 most frequent words and their frequency is ' + str(top_10))
```
